```python
positive = ["great","nice","awesome"]
negative = ["bad","worst","very bad"]
```

```python
sentence = input("Enter a string ")
posiCount = 0
negCount = 0
```

```python
for i in positive:
    if i in sentence:
        posiCount += 1
for i in negative:
    if i in sentence:
        negCount += 1        
```

```python
if posiCount > negCount:
    print("Positive")
elif negCount > posiCount:
    print("Negative")
else:
    print("Neutral")
```
